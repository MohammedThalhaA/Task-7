##Question No 1

###Write a Pandas program to get the powers of an array values element-wise. Note: First array elements raised to powers from second array
###Sample data: {'X':[78,85,96,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]}

###Expected Output:

###X Y Z

###0 78 84 86

###1 85 94 97

###2 96 89 96

###3 80 83 72

###4 86 86 83

```python
import pandas as pd

# Sample data
data = {'X': [78, 85, 96, 80, 86], 'Y': [84, 94, 89, 83, 86], 'Z': [86, 97, 96, 72, 83]}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame before applying the powers
print("Before applying powers:")
print(df)

# Create another DataFrame with the element-wise powers
df_powers = df.pow(df.index + 1, axis=0)

# Rename the columns based on the original DataFrame
df_powers.columns = df.columns

# Display the DataFrame after applying the powers
print("\nAfter applying powers:")
print(df_powers)
```

###This program uses the 'pow' method to raise each element of the DataFrame to the power of its row index + 1. The resulting DataFrame is then displayed.

### The 'axis=0' parameter in the 'pow' method specifies that the powers should be applied along the rows.

###exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura','Kevin', 'Jonas'],'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

```python
import pandas as pd
import numpy as np

# Sample Python dictionary data and list labels
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Create DataFrame with specified index labels
df = pd.DataFrame(exam_data, index=labels)

# Display the DataFrame
print(df)
```

###Use the diabetes data set from Pima Indians Diabetes , Perform Bivariate analysis

```python
pip install pandas matplotlib seaborn scikit-learn
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Pima Indians Diabetes dataset (replace 'path/to/diabetes.csv' with the actual path)
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
column_names = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
diabetes_df = pd.read_csv(url, names=column_names)

# Display the first few rows of the dataset
print(diabetes_df.head())

# Bivariate Analysis: Correlation Matrix
correlation_matrix = diabetes_df.corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title("Correlation Matrix of Pima Indians Diabetes Dataset")
plt.show()

# Bivariate Analysis: Scatter Plots
# Let's choose two variables for demonstration, e.g., Glucose vs. BMI
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Glucose", y="BMI", hue="Outcome", data=diabetes_df)
plt.title("Scatter Plot of Glucose vs. BMI with Outcome")
plt.show()
```

###Use the diabetes data set from Pima Indians Diabetes data set for performing the following:
###Apply Bivariate analysis:  Multiple Regression analysis

```python
pip install pandas statsmodels
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the Pima Indians Diabetes dataset (replace 'path/to/diabetes.csv' with the actual path)
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
column_names = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
diabetes_df = pd.read_csv(url, names=column_names)

# Split the data into independent variables (X) and the dependent variable (y)
X = diabetes_df.drop("Outcome", axis=1)  # Independent variables
X = sm.add_constant(X)  # Add a constant term for the intercept
y = diabetes_df["Outcome"]  # Dependent variable

# Perform multiple regression analysis
model = sm.OLS(y, X).fit()

# Display the summary of the regression analysis
print(model.summary())

```

##Question No 5

###Write a NumPy program to merge three given NumPy arrays of same shape

```python
import numpy as np

# Create three NumPy arrays of the same shape
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7, 8, 9], [10, 11, 12]])
array3 = np.array([[13, 14, 15], [16, 17, 18]])

# Merge the arrays along the first axis (axis=0)
merged_array = np.concatenate((array1, array2, array3), axis=0)

# Display the original arrays
print("Array 1:")
print(array1)

print("\nArray 2:")
print(array2)

print("\nArray 3:")
print(array3)

# Display the merged array
print("\nMerged Array:")
print(merged_array)
```
